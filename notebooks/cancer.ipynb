```python
import scipy.stats as stats
```

```python
prob_list = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean','radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst']
def plots(df,var):
    plt.subplot(1,3,1)
    plt.hist(df[var])
    plt.title('hist plot')
    
    plt.subplot(1,3,2)
    stats.probplot(df[var],dist='norm',plot=plt)
    plt.title('probplot')
    
    plt.subplot(1,3,3)
    sn.boxplot(df[var])
    plt.title('box_plot')
    
    plt.show()
```

```python
for i in prob_list:
    print(i)
    plots(df,i)
    
```

```python
score = []
for i in prob_list:
    score.append(stats.zscore(df[i]))
score
```

```python
data = df.copy()
upper = []
lower = []
for i in prob_list:
    upper.append(df[i].mean() + df[i].std()*3)
    lower.append(df[i].mean() - df[i].std()*3)
```

```python
j = 0
for i in prob_list:
    data[i] = data[(data[i]>lower[j]) & (data[i]<upper[j])];j+=1
```

```python
for i in prob_list:
    print(i)
    plots(df,i)
    
```

```python
use = ['radius_mean','perimeter_mean',
        'area_mean','concavity_mean',
        'concave points_mean','radius_worst','perimeter_worst', 'area_worst', 'concavity_worst']
remove_list = []
for i in df.columns:
    if i not in use:
        remove_list.append(i)
remove_list
```

```python
data = df.copy()

data = data.drop(remove_list,axis = 1)
data.head()
```

```python
data['result'] = df['diagnosis']
data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
test = data['result']
train = data.drop(['result'],axis = 1)
```

```python
rf = RandomForestClassifier()
rf.fit(train,test)
rf.score(train,test)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,test,test_size = 0.2)
```

```python
rf.fit(X_train,y_train)
rf.score(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
params = {
            'n_estimators':[int(x) for x in np.linspace(start=1,stop=500,num=10)],
            'max_depth':[int(x) for x in np.linspace(start=1,stop=30,num=5)],
            'min_samples_leaf':[2,5,10],
            'min_samples_split':[2,5,10],
            'criterion':['gini','entropy'],
}
```
